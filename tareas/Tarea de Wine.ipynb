# BASE DE DATOS
El conjunto de vinos estÃ¡ representado por la relaciÃ³n VINOS. Un vino  es  descrito  en la base de datos por un nÃºmero entero, un nombre, el aÃ±o de producciÃ³n y cuÃ¡ntos grados de alcohol posee. La clave primaria de la relaciÃ³n vinos es el atributo Num.

El conjunto de productores  estÃ¡ representado por  la relaciÃ³n PRODUCTORES. Un productor es descrito en la base de datos por un nÃºmero entero, su apellido, su nombre y la regiÃ³n que habita. La clave  de la relaciÃ³n es el atributo  Num.

Un productor puede producir cero o varios vinos. RecÃ­procamente, un vino es producido por cero o varios productores.

El conjunto de cosechas producidas estÃ¡ representado por la relaciÃ³n COSECHAS. Una tupla de COSECHAS representa una producciÃ³n particular de un vino (Nvin), realizada por un productor (Nprod) y especifica la cantidad de botellas producidas en esa cosecha.  La clave de la relaciÃ³n COSECHAS es el grupo  de atributos (Nvin, Nprod).

```python
wine=datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_dataframe["target"]=wine.target
x=wine_dataframe["alcohol"]
print(x.mean())
y=wine_dataframe["ash"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
alcohol=wine_dataframe["alcohol"]*50
ash=wine_dataframe["ash"]*50
```

```python
plt.scatter(x,y,c=col,s=100,alpha=0.5)
plt.xlabel("Alcohol(%)")
plt.ylabel("Ash(%)")
plt.title("Alcohol vs Ash")
plt.grid(True)
plt.show()
```
